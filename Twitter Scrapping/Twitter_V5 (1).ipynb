```python
# Approach 2 to access twitter metadata. 
#Reference: https://datascienceparichay.com/article/get-data-from-twitter-api-in-python-step-by-step-guide/

tweets_copy = []
for tweet in tweets:
    tweets_copy.append(tweet)
    
print("Total Tweets fetched:", len(tweets_copy))
```

```python
# Extracting only the required information and converting it into dataframe.

#intialize the dataframe
tweets_df = pd.DataFrame()
# populate the dataframe
for tweet in tweets_copy:
    hashtags = []
    medias = []
    try:
        for hashtag in tweet.entities["hashtags"]:
            hashtags.append(hashtag["text"])
        for media in tweet.entities["media"]:
            medias.append(media["media_url"])
        text = api.get_status(id=tweet.id, tweet_mode='extended').full_text
    except:
        pass
    tweets_df = tweets_df.append(pd.DataFrame({'id_str': tweet.id_str,
                                               'user_name': tweet.user.name, 
                                               'user_location': tweet.user.location,\
                                               'user_description': tweet.user.description,
                                               'user_verified': tweet.user.verified,
                                               'account_created_date': tweet.created_at,
                                               'user_follower_count': tweet.user.followers_count,
                                               'user_following_count': tweet.user.followers_count,
                                               'user_favourties_count_liftime': tweet.user.followers_count,
                                               'user_tweets_count_lifetime':tweet.user.statuses_count,
                                               'geo': tweet.geo,
                                               'coordinates':tweet.coordinates,
                                               'place':tweet.place,
                                             #  'retweet_count':tweet.retweet_count,
                                             #  'favorite_count':tweet.favorite_count,
                                               'language':tweet.lang,
                                               'text': text, 
                                               'hashtags': [hashtags if hashtags else None],
                                               'media_urls':[medias if medias else None],
                                               'source': tweet.source}))
    tweets_df = tweets_df.reset_index(drop=True)

tweets_df
```
